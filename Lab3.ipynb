## 1. Diferencia entre descenso de gradiente, descenso de gradiente por mini batches y descenso de gradiente estocástico

El **descenso de gradiente** es un algoritmo de optimización utilizado para minimizar funciones de pérdida en modelos de machine learning y deep learning. Se basa en calcular la dirección en la que la función de pérdida decrece más rápidamente y moverse en esa dirección mediante un factor denominado tasa de aprendizaje.

### Descenso de Gradiente Batch (GD - Gradient Descent)
**Descripción:**
- Utiliza todos los datos disponibles para calcular el gradiente y actualizar los parámetros en cada iteración.

**Ventajas:**
- Convergencia más estable y menos ruido en la actualización de los parámetros.
- Puede aprovechar operaciones vectorizadas para acelerar cálculos.

**Desventajas:**
- Requiere gran cantidad de memoria y es computacionalmente costoso para grandes volúmenes de datos.
- Puede ser lento debido al cálculo del gradiente sobre todo el dataset en cada iteración.

### Descenso de Gradiente Estocástico (SGD - Stochastic Gradient Descent)
**Descripción:**
- Calcula el gradiente y actualiza los parámetros utilizando un solo ejemplo de entrenamiento a la vez.

**Ventajas:**
- Requiere menos memoria y es computacionalmente más eficiente en grandes datasets.
- Puede escapar de óptimos locales debido a su naturaleza ruidosa.

**Desventajas:**
- Convergencia más ruidosa y menos estable.
- Requiere un ajuste más cuidadoso de la tasa de aprendizaje para evitar fluctuaciones excesivas.

### Descenso de Gradiente por Mini Batches (Mini-Batch GD)
**Descripción:**
- Divide el dataset en pequeños subconjuntos llamados mini batches y usa cada uno para calcular el gradiente y actualizar los parámetros.

**Ventajas:**
- Equilibra la eficiencia computacional y la estabilidad de la convergencia.
- Reduce el ruido comparado con SGD, pero sigue siendo más eficiente que el GD batch.
- Puede aprovechar la paralelización en hardware de cómputo acelerado (GPU/TPU).

**Desventajas:**
- Puede requerir ajuste fino del tamaño de los mini batches para lograr un buen equilibrio entre velocidad y estabilidad.

## 2. Comparación entre extracción de features (feature extraction) y selección de features (feature selection)

La **extracción de features** y la **selección de features** son dos técnicas en machine learning que buscan mejorar la eficiencia y el rendimiento de los modelos al reducir la dimensionalidad de los datos.

### Feature Extraction (Extracción de Features)
- Implica transformar los datos originales en un nuevo conjunto de features más representativos.
- Utiliza técnicas como PCA (Análisis de Componentes Principales), embeddings de palabras (Word2Vec, TF-IDF) y convoluciones en CNNs.
- **Ejemplo:** En visión por computadora, se pueden extraer características clave de imágenes usando redes convolucionales en lugar de utilizar píxeles individuales.

**Ventajas:**
- Permite representar los datos en un espacio reducido y más informativo.
- Reduce la redundancia y mejora la interpretabilidad de los datos.

**Desventajas:**
- Puede ser computacionalmente costoso.
- Puede perder información útil en la transformación si no se aplica correctamente.

### Feature Selection (Selección de Features)
- Consiste en elegir un subconjunto de las features originales sin transformación.
- Puede hacerse mediante técnicas como selección basada en correlación, algoritmos de búsqueda (forward selection, backward elimination) o técnicas de importancia de variables en modelos como Random Forest o LASSO.
- **Ejemplo:** En un modelo de predicción de enfermedades basado en datos clínicos, podríamos seleccionar únicamente las variables más relevantes (como presión arterial, colesterol y edad) y eliminar aquellas irrelevantes.

**Ventajas:**
- Reduce la dimensionalidad del modelo sin alterar la naturaleza de las features.
- Puede mejorar la interpretabilidad y reducir el overfitting.

**Desventajas:**
- No genera nuevas representaciones, lo que puede limitar el desempeño si las features originales no son adecuadas.
- Puede perder información si la selección no se realiza adecuadamente.

## 3. Arquitectura y funcionamiento de un Perceptrón de una sola capa

### Arquitectura:
Un perceptrón de una sola capa es un modelo de red neuronal simple que se basa en la combinación lineal de las entradas seguida por una función de activación. Está compuesto por:
- Una capa de entrada (neuronas de entrada con pesos asociados).
- Una única neurona de salida.
- Pesos y sesgo (bias) ajustables.
- Una función de activación (comúnmente la función escalón o la función sigmoide en algunas variantes).

### Funcionamiento:
1. **Cálculo de la suma ponderada:**
   - Cada entrada $x_i$ es multiplicada por un peso $w_i$, y luego se suma el sesgo $b$:
     
     $$ z = \sum w_i x_i + b $$

2. **Aplicación de la función de activación:**
   - Se pasa la suma ponderada a través de una función de activación, típicamente la función escalón:
     
     $$ f(z) = \begin{cases} 1, & \text{si } z \geq 0 \\ 0, & \text{si } z < 0 \end{cases} $$

3. **Clasificación de la salida:**
   - Si el resultado de la función de activación es 1, se clasifica en una categoría; si es 0, en la otra.

### Aprendizaje y Actualización de Parámetros:
El perceptrón aprende ajustando los pesos usando la **regla de aprendizaje del perceptrón**, que sigue estos pasos:
1. Se calcula la salida esperada y se compara con la salida obtenida.
2. Si la salida es incorrecta, se actualizan los pesos usando la ecuación:
   
   $$ w_i = w_i + \eta (y - \hat{y}) x_i $$
   
   donde:
   - $\eta$ es la tasa de aprendizaje.
   - $y$ es la salida deseada.
   - $\hat{y}$ es la salida obtenida.
   - $x_i$ es la entrada correspondiente.

Este proceso se repite iterativamente hasta que el modelo aprende a clasificar correctamente los datos linealmente separables.

### Limitaciones:
- Solo funciona para problemas linealmente separables.
- No puede manejar problemas más complejos, como los que requieren múltiples capas y funciones de activación más avanzadas (esto se soluciona con redes neuronales multicapa y backpropagation).

Implemente el descenso de gradiente estocástico, el descenso de gradiente y el descenso de gradiente por mini batches para una función polinómica de grado 3. Luego, grafique la función aproximada por cada uno de los métodos solicitados y la distribución real de puntos. 

Para esto considere que:

- El polinomio que debe usar es \(2x^3 - 3x^2 + 5x + 3\)
- Considere el uso de `np.polyval`
- Considere el uso de `np.polyder`
- Compare el tiempo de ejecución de cada uno de los métodos solicitados. De esto, mencione cuál método fue más rápido y por qué.
- Compare el fitness de cada método solicitado (recuerde hacer una evaluación numérica y no sólo visual). De esto mencione cuál método lo hizo mejor y por qué.


```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Definimos la función polinómica y su derivada
def polynomial(x):
    return 2*x**3 - 3*x**2 + 5*x + 3

def polynomial_derivative(x):
    return 6*x**2 - 6*x + 5

# Generamos datos de ejemplo
np.random.seed(42)
X = np.linspace(-2, 2, 100)
y = polynomial(X) + np.random.normal(0, 3, len(X))

# Función de costo MSE
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Descenso de Gradiente Estocástico
def sgd(X, y, lr=0.01, epochs=100):
    theta = np.random.randn()
    for epoch in range(epochs):
        for i in range(len(X)):
            grad = polynomial_derivative(X[i])
            theta -= lr * grad
    return theta

# Descenso de Gradiente Batch
def gradient_descent(X, y, lr=0.01, epochs=100):
    theta = np.random.randn()
    for epoch in range(epochs):
        grad = np.mean([polynomial_derivative(x) for x in X])
        theta -= lr * grad
    return theta

# Descenso de Gradiente Mini-Batch
def mini_batch_gd(X, y, lr=0.01, epochs=100, batch_size=10):
    theta = np.random.randn()
    for epoch in range(epochs):
        indices = np.random.permutation(len(X))
        for i in range(0, len(X), batch_size):
            batch_X = X[indices[i:i+batch_size]]
            grad = np.mean([polynomial_derivative(x) for x in batch_X])
            theta -= lr * grad
    return theta

# Comparación de tiempos de ejecución
start = time.time()
theta_sgd = sgd(X, y)
time_sgd = time.time() - start

start = time.time()
theta_gd = gradient_descent(X, y)
time_gd = time.time() - start

start = time.time()
theta_mbgd = mini_batch_gd(X, y)
time_mbgd = time.time() - start

# Evaluación del desempeño
fitness_sgd = mse(y, polynomial(X))
fitness_gd = mse(y, polynomial(X))
fitness_mbgd = mse(y, polynomial(X))

# Resultados
print(f"SGD: Theta = {theta_sgd}, Tiempo = {time_sgd:.4f} s, Fitness = {fitness_sgd:.4f}")
print(f"GD: Theta = {theta_gd}, Tiempo = {time_gd:.4f} s, Fitness = {fitness_gd:.4f}")
print(f"Mini-Batch GD: Theta = {theta_mbgd}, Tiempo = {time_mbgd:.4f} s, Fitness = {fitness_mbgd:.4f}")

# Generar gráficas individuales para cada modelo
methods = {
    "SGD": theta_sgd,
    "GD": theta_gd,
    "Mini-Batch GD": theta_mbgd
}

for method, theta in methods.items():
    plt.figure(figsize=(8, 5))
    plt.scatter(X, y, label="Datos reales", alpha=0.5)
    plt.plot(X, polynomial(X), label="Polinomio real", color='black', linestyle="dashed", linewidth=2)
    plt.plot(X, polynomial(X), label=f"{method}", linewidth=2)
    plt.legend()
    plt.title(f"Gradiente Descendente: {method}")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.show()
```
